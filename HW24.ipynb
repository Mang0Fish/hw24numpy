```python
#4
np.random.seed(80)
array4 = np.random.randint(1,101,(4,4))
print(array4)
print(np.sum(array4, axis = 1))
print(np.sum(array4, axis = 0))
print(np.max(array4, axis = 1))
print(np.max(array4, axis = 0))
print(np.min(array4, axis = 1))
print(np.min(array4, axis = 0))
print(np.mean(array4, axis = 1))
print(np.mean(array4, axis = 0))
print(array4[array4 > 50])
#Median is the number the is in the middle of an ascending numbers group,
#for example 3 is the median in [1,2,3,8,10]
#Standard deviation is the root of Variance,
#with variance we calculate the squared distance of each number 
#in the group from its mean, for example
#in [8,10,10,12,14] we first calculate its average which is 10.8
#then we subtract the mean from each member and then square it
#then after we squared those numbers, we calculate their mean,
#and this mean is Variance, its root is called Standard Deviation.
```
